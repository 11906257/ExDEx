### Dataset "Farm structure" of EUROSTAT


```python
df_agricultural_holdings_by_agricultural_area = pd.read_csv('EUROSTAT/Agricultural_Holdings_By_Agricultural_Area.tsv', sep='\t')
df_farm_land_use = pd.read_csv('EUROSTAT/Farm_Land_Use.tsv', sep='\t')
df_agricultural_holdings_with_livestock = pd.read_csv('EUROSTAT/Agricultural_Holdings_With_Livestock.tsv', sep='\t')
df_agricultural_holdings_with_organic_area = pd.read_csv('EUROSTAT/Agricultural_Holdings_With_Organic_Area.tsv', sep='\t')
df_agricultural_holdings_by_legal_form = pd.read_csv('EUROSTAT/Agricultural_Holdings_By_Legal_Form.tsv', sep='\t')
df_farm_labour_force = pd.read_csv('EUROSTAT/Farm_Labour_Force.tsv', sep='\t')
df_agricultural_holdings_by_holder_age = pd.read_csv('EUROSTAT/Agricultural_Holdings_By_Holder_Age.tsv', sep='\t')
df_agricultural_holdings_by_manager_age = pd.read_csv('EUROSTAT/Agricultural_Holdings_By_Manager_Age.tsv', sep='\t')
df_agricultural_holdings_by_manager_training = pd.read_csv('EUROSTAT/Agricultural_Holdings_By_Manager_Training.tsv', sep='\t')
df_livestock_density_index = pd.read_csv('EUROSTAT/Livestock_Density_Index.tsv', sep='\t')
```

```python
plt.figure(figsize=(12, 6))
eu_data = df_agricultural_holdings_by_agricultural_area[df_agricultural_holdings_by_agricultural_area['freq,uaarea,unit,geo\TIME_PERIOD'] == 'A,TOTAL,HLD,EU27_2020'].iloc[0]

if not eu_data.empty:
    years = [int(col.strip().replace(' ', '')) for col in df_agricultural_holdings_by_agricultural_area.columns 
         if col != 'Country' and str(col).strip().replace(' ', '').isdigit()]
    values = [int(str(v).strip().replace(' ', '')) for v in eu_data.values if str(v).strip().replace(' ', '').isdigit()]
    plt.plot(years[1:len(years)-1], values, marker='o', label="Agricultural Holdings")

plt.title('EU - Agricultural holdings by agricultural area Over Years')
plt.xlabel('Year')
plt.ylabel('Holdings')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
